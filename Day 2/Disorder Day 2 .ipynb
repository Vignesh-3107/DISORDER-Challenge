```python
import pandas as pd
```

```python
df=pd.read_csv('Road accident data (11-19) .csv')
```

```python
df.head()
```

```python
datasets=df.drop(['Unnamed: 1'],axis=1)
```

```python
datasets.head()
```

```python
datasets.isnull().sum()
```

```python
datasets['Country'].nunique()
```

```python
dataset=datasets.dropna()
dataset.isnull().sum()
```

## TOP 10 COUNTRY IN 2019 ROAD ACCIDENTS :

```python
data=[]
for C in range(0,dataset['Country'].nunique()):
    data.append([dataset['Country'][C],int(dataset['2019'][C])])
data1=pd.DataFrame(data,columns = ['Country','2019']).sort_values(by='2019', ascending=False).head(10)
```

```python
data1.plot(x = 'Country', y = '2019',  kind = 'barh', figsize = (20,10),color='red')
```
